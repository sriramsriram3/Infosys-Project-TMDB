# Unique values of the columns

**Finding unique values in the columns**

```python

def unique_values_in_col(col):
  unique_values=col.str.split(',').explode().unique()     # split the data into rows
  unique_values=[val.strip() for val in unique_values]
  unique_values=list(set(unique_values))
  unique_values=[val for val in unique_values if val]    #to perform filtering --> ''

  return unique_values
```

# One Hot Encoding

**One Hot Encoding**

```python
def one_hot_encoding(df,column,unique_list):
  #creating column names which are in unique_list
  encoded_cols=[f"{value}" for value in unique_list]
  #creating a new dataframe
  encoded_df=pd.DataFrame(columns=encoded_cols)

  for index,row in df.iterrows():
    col_values= row[column].split(',')
    for value in col_values:
      value=value.strip()
      encoded_df.loc[index,f"{value}"]=1
    encoded_df.fillna(0,inplace=True)
  encoded_df=encoded_df.astype(int)
  df.drop(columns=column,inplace=True)
  df=pd.concat([df,encoded_df],axis=1)
  return df
```
