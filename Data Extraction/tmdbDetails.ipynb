```python
import requests
import pandas as pd

# Replace with your TMDB API key
api_key = "b95c9fdd84a97ae3467206de8b62cfe7"

# Base URL for movie details
base_url = f"https://api.themoviedb.org/3/movie/{{movie_id}}?api_key={api_key}"

# Year to search
year = 2019

# Desired fields
desired_fields = [
    "id",
    "budget",
    "genres",
    "homepage",
    "imdb_id",
    "original_language",
    "original_title",
    "overview",
    "popularity",
    "poster_path",
    "production_companies",
    "production_countries",
    "release_date",
    "runtime",
    "spoken_languages",
    "status",
    "tagline",
    "title",
]

# Function to fetch movie details
def fetch_movie_details(movie_id):
    url = base_url.format(movie_id=movie_id)
    response = requests.get(url)
    data = response.json()
    # Extract relevant fields
    movie_data = {field: data.get(field) for field in desired_fields}
    # Process nested structures (e.g., genres, production companies)
    movie_data["genres"] = [genre["name"] for genre in data.get("genres", [])]
    movie_data["production_companies"] = [
        company["name"] for company in data.get("production_companies", [])
    ]
    movie_data["production_countries"] = [
        country["name"] for country in data.get("production_countries", [])
    ]
    return movie_data

# List of movie IDs (you can replace with your source)
movie_ids = data_list  # Replace with actual IDs

# Collect movie data with budget check and limit
movies_data = []
count = 0  # Counter to track collected movies
for movie_id in movie_ids:
    try:
        movie_data = fetch_movie_details(movie_id)
        # Include movie data only if budget is greater than zero and limit reached
        if movie_data.get("budget", 0) > 0 and count < 100:
            movies_data.append(movie_data)
            count += 1
    except Exception as e:
        print(f"Error fetching movie ID {movie_id}: {e}")

# Create pandas DataFrame and export to CSV (handle empty list case)
if movies_data:
    df = pd.DataFrame(movies_data)
    df.to_csv("movie_data_2019_budget_limited.csv", index=False)
    print(f"Collected {count} movies with budget > 0. Saved to movie_data_2019_budget_limited.csv")
else:
    print("No movies found with budget > 0 in the provided list or API quota reached.")
```
